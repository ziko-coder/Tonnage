```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sheet_name='../data/precoce_all.csv'
```

```python
df = pd.read_csv(sheet_name)
```

# Stats for branches

```python
df_ab = df[['DChar3', 'DChar4', 'DChar5']]
missing_values = df_ab.isnull().sum()
```

```python
plt.figure(figsize=((25,10)))

plt.subplot(1,2,1)
total_missing = df_ab.isnull().sum().sum()
plt.pie(missing_values, labels=['DChar3', 'DChar4', 'DChar5'], autopct='%1.1f%%')
plt.title(f'Total Missing Values: {total_missing}')

plt.subplot(1,2,2)
missing_values = df_ab.isnull().sum()
missing_values.plot(kind='bar')
plt.xlabel('Columns')
plt.ylabel('Number of missing values')
plt.title('Bar Chart of Missing Values by Column')


plt.show()
```

```python
plt.figure(figsize=((25,10)))
total_missing = df_ab.isnull().sum().sum()
plt.pie(missing_values, labels=['DChar3', 'DChar4', 'DChar5'], autopct='%1.1f%%')
plt.title(f'Total Missing Values: {total_missing}')
plt.show()
```

```python
sns.heatmap(df[['DChar3', 'DChar4', 'DChar5']].isnull(), cmap='Blues', cbar=False) # missing values in blue

plt.xlabel('Columns')
plt.title('Heatmap of Missing Values by Column')

plt.show()
```
